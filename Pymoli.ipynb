```python
import pandas as pd
file = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file)
```

```python
#creating function for percentage 
def percentage(part, whole):
  return 100 * float(part)/float(whole)
```

```python
total_players = pd.DataFrame({"Total Player": [0]})
```

```python
purchase_data.head()
```

#### Player Count
* Display the total number of players

```python
# Player Count
total_players["Total Player"] = purchase_data["Purchase ID"].count()
total_players.head()
```

#### Purchasing Analysis (Total)
* Displays "Number of Unique Items", "Average Price", "Number of Purchases" and "Total Revenue"

```python
#Purchasing Analysis
columns = {
    "Number of Unique Items": [0],
    "Average Price": [0],
    "Number of Purchases": [0],
    "Total Revenue": [0]
}
purchasing_analysis = pd.DataFrame(columns)
purchasing_analysis["Number of Unique Items"] = purchase_data["Item ID"].nunique()
purchasing_analysis["Average Price"] = purchase_data["Price"].mean()
purchasing_analysis["Number of Purchases"] = purchase_data["Purchase ID"].count()
purchasing_analysis["Total Revenue"] = purchase_data["Price"].sum()
purchasing_analysis.head()
```

#### Gender Demographics 
* Displays Male, Female and Other/Non-Disclosed with Total Count and Percentage of Players

```python
demographic_data = {
    "Total Count": [0],
    "Percentage of Players": [0]
}
#creating data frame of gender values
gender_count = purchase_data["Gender"].value_counts()
#creating gender data frame
gender_df = pd.DataFrame(demographic_data, index=["Male", "Female", "Other / Non-Disclosed"])

# adding male, female, other/non-disclosed values
gender_df["Total Count"][0] = gender_count["Male"]
gender_df["Total Count"][1] = gender_count["Female"]
gender_df["Total Count"][2] = gender_count["Other / Non-Disclosed"]

# calculating percentage for males
male_percent = percentage(gender_count["Male"], purchase_data["Gender"].count())
gender_df["Percentage of Players"][0] = male_percent
# calculating percentage for females
female_percent = percentage(gender_count["Female"], purchase_data["Gender"].count())
gender_df["Percentage of Players"][1] = female_percent
# calculating percentage for other/non-disclosed
other_non_disclosed = percentage(gender_count["Other / Non-Disclosed"], purchase_data["Gender"].count())
gender_df["Percentage of Players"][2] = other_non_disclosed


gender_df.head()
```

```python
gender_purchase = {
    "Purchase Count": [0],
    "Average Purchase Price": [0],
    "Total Purchase Value": [0],
    "Avg Total Purchase per Person": [0]
}

gender_purchases = pd.DataFrame(gender_purchase, index=["Male", "Female", "Other / Non-Disclosed"])

purchases = purchase_data[["Gender", "Price"]]
male_purchases = purchases[purchases.Gender == 'Male']
female_purchases = purchases[purchases.Gender == 'Female']
other_purchases = purchases[purchases.Gender == 'Other / Non-Disclosed']
#
gender_purchases["Purchase Count"][0] = gender_count["Male"]
gender_purchases["Purchase Count"][1] = gender_count["Female"]
gender_purchases["Purchase Count"][2] = gender_count["Other / Non-Disclosed"]

#Average Purchase Price
gender_purchases["Average Purchase Price"][0] = male_purchases["Price"].mean()
gender_purchases["Average Purchase Price"][1] = female_purchases["Price"].mean()
gender_purchases["Average Purchase Price"][2] = other_purchases["Price"].mean()

#Total Purchase Value
gender_purchases["Total Purchase Value"][0] = male_purchases["Price"].sum()
gender_purchases["Total Purchase Value"][1] = female_purchases["Price"].sum()
gender_purchases["Total Purchase Value"][2] = other_purchases["Price"].sum()

#Average Total Purchase per Person
gender_purchases["Avg Total Purchase per Person"][0] = male_purchases["Price"].mean()
gender_purchases["Avg Total Purchase per Person"][1] = female_purchases["Price"].mean()
gender_purchases["Avg Total Purchase per Person"][2] = other_purchases["Price"].mean()

gender_purchases
```
